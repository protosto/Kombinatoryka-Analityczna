```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def function_1(z):
    return np.abs(z)

def function_2(z):
    return np.abs(z**2)

def function_3(z):
    return np.abs(1 / (2 - np.exp(z)))

def function_4(z):
    return np.abs(1 / ((1 - z) * (1 - z/2) * (1 - z/3)))

def function_5(z):
    numerator = 1 + z + z**2
    denominator = 1 - z - z**2 - z**3
    return np.abs(numerator / denominator)

def function_6(z):
    return np.abs(np.exp(-z) / (2 - np.exp(z)))

def function_7(z):
    return np.abs(np.sqrt(1 - 4 * z))

# Tworzenie siatki punktów z i wartości funkcji na tej siatce
z_real = np.linspace(-5, 5, 100)
z_imag = np.linspace(-5, 5, 100)
z_real, z_imag = np.meshgrid(z_real, z_imag)
z = z_real + 1j * z_imag

# Obliczanie wartości funkcji na siatce punktów
result_1 = function_1(z)
result_2 = function_2(z)
result_3 = function_3(z)
result_4 = function_4(z)
result_5 = function_5(z)
result_6 = function_6(z)
result_7 = function_7(z)

# Tworzenie interaktywnych wykresów 3D
fig = plt.figure(figsize=(18, 12))

ax1 = fig.add_subplot(231, projection='3d')
ax1.plot_surface(z_real, z_imag, np.real(result_1), cmap='viridis')
ax1.set_title('f(z) = |z|')

ax2 = fig.add_subplot(232, projection='3d')
ax2.plot_surface(z_real, z_imag, np.real(result_2), cmap='viridis')
ax2.set_title('f(z) = |z^2|')

ax3 = fig.add_subplot(233, projection='3d')
ax3.plot_surface(z_real, z_imag, np.real(result_3), cmap='viridis')
ax3.set_title('f(z) = |1 / (2 - e^z)|')

ax4 = fig.add_subplot(234, projection='3d')
ax4.plot_surface(z_real, z_imag, np.real(result_4), cmap='viridis')
ax4.set_title('f(z) = |1 / ((1 - z) * (1 - z/2) * (1 - z/3))|')

ax5 = fig.add_subplot(235, projection='3d')
ax5.plot_surface(z_real, z_imag, np.real(result_5), cmap='viridis')
ax5.set_title('f(z) = |(1 + z + z^2) / (1 - z - z^2 - z^3)|')

ax6 = fig.add_subplot(236, projection='3d')
ax6.plot_surface(z_real, z_imag, np.real(result_6), cmap='viridis')
ax6.set_title('f(z) = |e^(-z) / (2 - e^z)|')

# Wykres 3D dla funkcji 7
fig_7 = plt.figure(figsize=(8, 6))
ax7 = fig_7.add_subplot(111, projection='3d')
ax7.plot_surface(z_real, z_imag, np.real(result_7), cmap='viridis')
ax7.set_title('f(z) = |sqrt(1 - 4z)|')

plt.show()
```
