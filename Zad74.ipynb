```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def parametric_equation_1(r, t):
    x = np.real(r * np.exp(1j * t))
    y = np.imag(r * np.exp(1j * t))
    z = np.real(np.sqrt(r) * np.exp(1j * t / 2))
    return x, y, z

def parametric_equation_2(r, t):
    x = np.real(r * np.exp(1j * t))
    y = np.imag(r * np.exp(1j * t))
    z = np.imag(np.sqrt(r) * np.exp(1j * t / 2))
    return x, y, z

def parametric_equation_3(r, t):
    x = np.real(r * np.exp(1j * t))
    y = np.imag(r * np.exp(1j * t))
    z = np.abs(np.log(r) + 1j * t)
    return x, y, z

# Tworzenie siatki punktów dla parametrów r i t
r_1 = np.linspace(0, 1, 100)
t_1 = np.linspace(0, 4 * np.pi, 100)
r_1, t_1 = np.meshgrid(r_1, t_1)

r_2 = np.linspace(0, 1, 100)
t_2 = np.linspace(0, 4 * np.pi, 100)
r_2, t_2 = np.meshgrid(r_2, t_2)

r_3 = np.linspace(0.1, 10, 100)
t_3 = np.linspace(0, 8 * np.pi, 100)
r_3, t_3 = np.meshgrid(r_3, t_3)

# Wygenerowanie punktów dla każdej funkcji parametrycznej
points_1 = parametric_equation_1(r_1, t_1)
points_2 = parametric_equation_2(r_2, t_2)
points_3 = parametric_equation_3(r_3, t_3)

# Tworzenie wykresów 3D
fig_1 = plt.figure(figsize=(8, 8))
ax_1 = fig_1.add_subplot(111, projection='3d')
ax_1.plot_surface(points_1[0], points_1[1], points_1[2], cmap='viridis')
ax_1.set_title('Sr')

fig_2 = plt.figure(figsize=(8, 8))
ax_2 = fig_2.add_subplot(111, projection='3d')
ax_2.plot_surface(points_2[0], points_2[1], points_2[2], cmap='viridis')
ax_2.set_title('Sc')

fig_3 = plt.figure(figsize=(8, 8))
ax_3 = fig_3.add_subplot(111, projection='3d')
ax_3.plot_surface(points_3[0], points_3[1], points_3[2], cmap='viridis')
ax_3.set_title('L')

plt.show()
```
